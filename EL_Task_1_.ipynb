## **3) Converting Categorical Features into numerical using Label Encoding**

```python
data.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder

cols_to_encode = ['Sex', 'Embarked', 'Cabin', 'Ticket']

le = LabelEncoder()

for col in cols_to_encode:
    data[col] = le.fit_transform(data[col])

print(data[cols_to_encode].head())
print(data[cols_to_encode].dtypes)
```

## **4) Normalize/standardize the numerical features.**

```python
data
```

```python
# Selecting numerical columns
num_features = ['Age', 'Fare', 'SibSp', 'Parch']
data[num_features].head(2)
```

```python
#Standardization (Z-score normalization)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data[num_features] = scaler.fit_transform(data[num_features])

data[num_features].describe().round(2)

```

## **5)Visualize outliers using boxplots and remove them.**

```python
plt.figure(figsize=(12, 6))
for i, col in enumerate(num_features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=data[col], color='skyblue')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

### Remove outliers using the IQR method

```python
# Defining a function to remove outliers

def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data
```

```python
# Applying it to each numeric feature.
for col in num_features:
    data = remove_outliers_iqr(data, col)
```

```python
# After removing outlier
plt.figure(figsize=(12, 6))
for i, col in enumerate(num_features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=data[col], color='lightgreen')
    plt.title(f'Boxplot of {col} (After Removing Outliers)')
plt.tight_layout()
plt.show()
```
